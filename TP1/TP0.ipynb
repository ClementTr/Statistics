## Question 6

```python
from timeit import default_timer as timer
import matplotlib.pyplot as plt
from numpy import *
import numpy as np

iteration_number = [10, 1000, 100000, 10000000]

# With a loop #
time_array1 = []
for iteration in iteration_number:
    start1 = timer()
    result1 = 4/3
    for k in range(2, iteration):
        result1 = result1*((4*(k**2))/((4*(k**2))-1))
        result1 = result1*2
    end1 = timer()
    time1 = end1 - start1
    time_array1.append(time1)

    
# With numpy #
time_array2 = []
for iteration in iteration_number:
    start2 = timer()
    arr = arange(1, iteration)
    all_arr = (4*(arr**2))/((4*(arr**2))-1)
    result2 = np.prod(all_arr)
    result2 = np.multiply(result2, 2)
    end2 = timer()
    time2 = end2 - start2
    time_array2.append(time2)
    
# Visualization #
plt.plot(iteration_number, time_array1, 'k')
plt.plot(iteration_number, time_array2, 'r')
plt.xlabel('Iterations')
plt.ylabel('Time')
plt.legend(['timing with loop', 'timing with numpy'], loc='upper left')
plt.show()
```

## Question 7

```python
def quicksort(l1):
    """ a sorting algorithm with a pivot value"""
    if len(l1) <= 1:
        return l1
    else:
        pivot = l1[len(l1)-1]
        less = []
        greater = []
        for x in l1:
            if x <= pivot:
                less.append(x)
            else:
                greater.append(x)
        less.extend(greater)
        return less
    
my_list = [-2, 3, 5, 1, 3]
print(my_list)
print(quicksort(my_list))
```

## Question 8

```python
# Generate matrix size(5,6) between -1 and 1 #
M = 2 * np.random.random_sample((5, 6)) - 1
print(M)

print("\n\n")

# Matrix computation #
for i in range(0, M.shape[1], 2):
    M[:,i] = M[:,i]-2*M[:,i+1]
print(M)

print("\n\n")

# Replace negative values #
np.where(M < 0, 0, M)
print(M)
```
